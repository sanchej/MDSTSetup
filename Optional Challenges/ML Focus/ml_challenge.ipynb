```python
# you are free to use any package you deem fit
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import sklearn.linear_model as linmod
```

## Dataset

```python
# DO NOT MODIFY
from sklearn.datasets import make_moons

X, Y = make_moons(random_state=42, n_samples=(50, 450), noise=0.25)
```

```python
# LinearRegression(), BayesianRidge(), Ridge(), Lasso(), LassoLars()
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=.3, random_state=0)
```

## Training

Training all linear models:

```python
model1 = linmod.LinearRegression()
model2 = linmod.LinearRegression(fit_intercept=False)
model3 = linmod.BayesianRidge()
model4 = linmod.Ridge(alpha= 0.5)
model5 = linmod.RidgeCV(alphas=np.logspace(-6, 6, 13))
model6 = linmod.Lasso(alpha= 0.5)
model7 = linmod.LassoLars(alpha= 0.5)
model8 = linmod.TweedieRegressor(power=1, alpha=0.5, link='log')
```

```python
model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model3.fit(X_train, y_train)
model4.fit(X_train, y_train)
model5.fit(X_train, y_train)
model6.fit(X_train, y_train)
model7.fit(X_train, y_train)
model8.fit(X_train, y_train)
```

## Testing / Optimization

The following code block encodes the results to fit the format of the results(either 1 or 0)

```python
res1 = model1.predict(X_test)
res1 = [1 if x > 0.5 else 0 for x in res1]
res2 = model2.predict(X_test)
res2 = [1 if x > 0.5 else 0 for x in res2]
res3 = model3.predict(X_test)
res3 = [1 if x > 0.5 else 0 for x in res3]
res4 = model4.predict(X_test)
res4 = [1 if x > 0.5 else 0 for x in res4]
res5 = model5.predict(X_test)
res5 = [1 if x > 0.5 else 0 for x in res5]
res6 = model6.predict(X_test)
res6 = [1 if x > 0.5 else 0 for x in res6]
res7 = model7.predict(X_test)
res7 = [1 if x > 0.5 else 0 for x in res7]
res8 = model8.predict(X_test)
res8 = [1 if x > 0.5 else 0 for x in res8]
```

```python
print(model1.score(X_test, y_test))
print(model2.score(X_test, y_test))
print(model3.score(X_test, y_test))
print(model4.score(X_test, y_test))
print(model5.score(X_test, y_test))
print(model6.score(X_test, y_test))
print(model7.score(X_test, y_test))
print(model8.score(X_test, y_test))
```

The code block below displays the error results. For the model errors, root mean squared error was used to penalize large errors. In the case of the error among the encoded results, a simple error was used since the manipulated data would have innacurately increased the root mean squared error of the encoded results.

```python
print("Model 1 error: " + str(np.mean((model1.predict(X_test)-y_test)**2)**0.5) + " Encoding error: " + str(np.mean((res1-y_test))))
print("Model 2 error: " + str(np.mean((model2.predict(X_test)-y_test)**2)**0.5) + " Encoding error: " + str(np.mean((res2-y_test))))
print("Model 3 error: " + str(np.mean((model3.predict(X_test)-y_test)**2)**0.5) + " Encoding error: " + str(np.mean((res3-y_test))))
print("Model 4 error: " + str(np.mean((model4.predict(X_test)-y_test)**2)**0.5) + " Encoding error: " + str(np.mean((res4-y_test))))
print("Model 5 error: " + str(np.mean((model5.predict(X_test)-y_test)**2)**0.5) + " Encoding error: " + str(np.mean((res5-y_test))))
print("Model 6 error: " + str(np.mean((model6.predict(X_test)-y_test)**2)**0.5) + " Encoding error: " + str(np.mean((res6-y_test))))
print("Model 7 error: " + str(np.mean((model7.predict(X_test)-y_test)**2)**0.5) + " Encoding error: " + str(np.mean((res7-y_test))))
print("Model 8 error: " + str(np.mean((model8.predict(X_test)-y_test)**2)**0.5) + " Encoding error: " + str(np.mean((res8-y_test))))
```

Now comparing the metrics of score and average error squared, it is clear that the LinearRegression() model has the best results in regard to error. In regards to score, the score typically indicates the linearity of the model and when tested on models 6 and 7 (Lasso and LassoLars) it is clear when graphed that while the predicted testing is extremely linear, the accuracy of the predictions are lacking in comparison to the other models. This can be seen by the code block below. When visualizing the testing results of the linear regression model, it is evident that the predicted data points are being influenced by the lower training data values towards the right half of the graph. The results were encoded to aggregate datapoints closer to the associated binary value that each datapoint was closest to. Doing so, gave the linear regression model the best error score yet again, and when graphed showed the data to be significantly more accurate. Unfortunately, not too many of the outliers were adequately guessed but this could potentially change by changing the encoding from values on either side of 0.5 to perhaps using a number larger than 0.5 but still under 1. It is also likely that more outliers would cause the model to better adapt. Overall, while the accuracy is not the best, it is somewhat decent and when encoding is utilized, it has only an error of about 10.67% which is not terrible.
