**Background**:

There are exactly 100 sneaker shops on a sneaker retailing website, and each of these shops sells only one model of shoe. We want to do some analysis of the average order value (AOV). When we look at orders data over a 30 day window, we naively calculate an AOV of $3145.13. Given that we know these shops are selling sneakers, a relatively affordable item, something seems wrong with our analysis. 

**Questions**:

- What went wrong with this metric (AOV) and the current analysis in general. Support your answers with appropriate visualizations.

- Propose a new metric that better represents the behavior of the stores' customers. Why are these metrics better and what are their values?

- Using the `created_at` column as a proxy for the date and time of order creation, provide a brief analysis of the trends in the customers' purchasing behavior.

Show all of your work in this notebook.

```python
# Imports and csv file reading
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.mode.chained_assignment = None
df = pd.read_csv('./orders.csv')
```

The code block below adds a date and time column to the dataframe for later use. The third to last line which is commented out (line 6) provided a view of expensive orders which I believed would be above $2000 in value. Based on this view I noticed some purchases had purchased a total of 2000 items which seemed highly unusual. By applying ".describe()" to the dataframe, I noticed that 2000 items was in fact the maximum total items ordered and the final line in the code helped me discover that 17 items in the data set had these extreme outliers that likely skewed the data.

```python
df['date'] = df['created_at'].str[5:10]
df['time'] = df['created_at'].str[11:13]
for i, row in df.iterrows():
    if row["time"][1] == ":":
        df.at[i,"time"] = "0" + row["time"][0]
# df[df["order_value"] > 2000]
# df[df["order_value"] > 2000].describe()
df[df["total_items"] == 2000].count()
```

The code block below shows a preliminary line chart of the average order values per date. Based on the scale, it confirms the AOV is highly skewed.

```python
df.groupby('date')["order_value"].mean().plot()
plt.ylabel('AOV')
plt.title('AOV Over Time')
```

The code block below creates a new column in the data frame which indicates if the order is an outlier (when 2000 items are ordered). The problematic order entries are then filtered out and the subsequent data is plotted.

```python
df["max_items"] = df["total_items"] == 2000
df[df["max_items"] == False].groupby('date')["order_value"].mean().plot()
plt.ylabel('AOV')
plt.title('Max Item Entries Removed')
```

Because the graph above still has some extreme values, the code block below checks the average order value of orders under $1000. This confirms a more reasonable average of around $300 which is vastly different from the graph above.

```python
df[df["order_value"] < 1000].groupby('date')["order_value"].mean().plot()
plt.ylabel('AOV')
plt.title('AOV of values under 1000')
```

The code block below displays entries that do not purchase the maximum amount of items but still have purchase values of above $1000. Immediately, I noticed a pattern in the value to items ratio and sure enough, many entries had the same unreasonable price to items ratio of 25725:1.

```python
df.loc[(df.loc[:, 'max_items'] == False) & (df.loc[:, 'order_value'] > 1000)].groupby('date').head()  
```

The code block below analyzes a common trait of the two outliers of extremely high purchases. Upon further analysis, the values with excessive price to items ratios all came from the same shop as seen in the table above. Looking back at the entries that have purchased 2000 items, these purchases all come from the same user. Now while it is possible that these purchases are legitimate, it is clear that the shop with a "shop_id" of 78 has significantly expensive items, and the user with a "user_id" of 607 buys an abnormally excessive amount of items. The code block below shows this to be true throughout the data.

```python
# df[df["shop_id"] == 78]
df[df["user_id"] == 607]
```

With the discoveries made prior to this point, I decided to make another column to record the purchases with the excessive prices of $25725 per item. Then after removing the outlier conditions, the adjusted average order value over time plot is created.

```python
df["max_price"] = df["order_value"] / df["total_items"] == 25725
df.loc[(df.loc[:, 'max_items'] == False) & (df.loc[:, 'max_price'] == False)].groupby('date')["order_value"].mean().plot()
plt.ylabel('AOV')
plt.title('Adjusted AOV Over Time')
```

The code block below gets an overall average order value and average total items which fall much more in line with reasonable prices and amounts.

```python
AOV = df.loc[(df.loc[:, 'max_items'] == False) & (df.loc[:, 'max_price'] == False)]["order_value"].mean()
ATI = df.loc[(df.loc[:, 'max_items'] == False) & (df.loc[:, 'max_price'] == False)]["total_items"].mean()
print("Average Order Value (Adjusted and Rounded): " + str(AOV.round(2)))
print("Average Total Items (Adjusted and Rounded): " + str(ATI.round()))
```

Just out of curiosity I wanted to check the adjusted average order values per hour of the day. While no significant trends are apparent, it may shed some light on peak order hours.

```python
df.loc[(df.loc[:, 'max_items'] == False) & (df.loc[:, 'max_price'] == False)].groupby('time')["order_value"].mean().plot()
plt.ylabel('AOV')
plt.xlabel('Hour')
plt.title('Adjusted AOV Over Time of Day')
```

Finally, analyzing the data without any outliers removed shows that 04:00:00 is the most profitable time for purchases which is likely due to the user with "user_id" 607 as they purchase at the 04:00:00 for every purchase. If I had to guess this purchase probably refers to a mass order from a store perhaps. Therefore, when accounting for user 607, the most profitable time is 04:00:00 and based on the code block after which discounts user 607, the most profitable time is 05:00:00.

```python
print("Max AOV over each hour: " + str(df.groupby('time')["order_value"].mean().max()))
df.groupby('time')["order_value"].mean()
```
